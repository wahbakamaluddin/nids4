```python
# Modify column names and drop unnecessary columns
df1 = df1[list(final_cols_df1.keys())].rename(columns=final_cols_df1)
df2 = df2[list(final_cols_df2.keys())].rename(columns=final_cols_df2)
df3 = df3[list(final_cols_df1.keys())].rename(columns=final_cols_df1)
```

```python
df3['Protocol'] = df3['Protocol'].replace({'TCP': 6, 'UDP': 17, 'ICMP': 1}).astype(int)
```

```python
# Comparing dataframes
# Generate flow key
df1['flow_key'] = (
    df1['Source IP'].astype(str) + ':' +
    df1['Source Port'].astype(str) + '-' +
    df1['Destination IP'].astype(str) + ':' +
    df1['Destination Port'].astype(str) + '-' +
    df1['Protocol'].astype(str)
    )

df2['flow_key'] = (
    df2['Source IP'].astype(str) + ':' +
    df2['Source Port'].astype(str) + '-' +
    df2['Destination IP'].astype(str) + ':' +
    df2['Destination Port'].astype(str) + '-' +
    df2['Protocol'].astype(str)
    )

df3['flow_key'] = (
    df3['Source IP'].astype(str) + ':' +
    df3['Source Port'].astype(str) + '-' +
    df3['Destination IP'].astype(str) + ':' +
    df3['Destination Port'].astype(str) + '-' +
    df3['Protocol'].astype(str)
    )
```

```python
# Find intersections
print(df1['flow_key'].head())
print(df2['flow_key'].head())
print(df3['flow_key'].head())
```

```python
matched_keys_12 = set(df1['flow_key']).intersection(set(df2['flow_key']))
matched_keys_13 = set(df1['flow_key']).intersection(set(df3['flow_key']))
matched_keys_23 = set(df2['flow_key']).intersection(set(df3['flow_key']))

print(f"Number of matched flows between df1 and df2: {len(matched_keys_12)}")
print(f"Number of matched flows between df1 and df3: {len(matched_keys_13)}")
print(f"Number of matched flows between df2 and df3: {len(matched_keys_23)}")
```

```python
# Remove non-matching flows from df1
df1 = df1[df1['flow_key'].isin(matched_keys_12)]
```

```python
# Compare number of rows
print(f"Number of rows (df1): {len(df1)}")
print(f"Number of rows (df2): {len(df2)}")
print(f"Number of rows (df3): {len(df3)}")
```
