```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('/home/wahba/Documents/nids5/test/dataset/cicids2017_data_preprocessing2.csv')
```

```python
attack_counts = df['Label'].value_counts()
percentages = (attack_counts / attack_counts.sum() * 100).round(2)

threshold = 0.005
percentages = attack_counts / attack_counts.sum()
small_slices = percentages[percentages < threshold].index.tolist()
attack_counts['Other'] = attack_counts[small_slices].sum()
attack_counts.drop(small_slices, inplace = True)

sns.set_palette('pastel')
plt.figure(figsize = (8, 8))
plt.pie(attack_counts.values, labels = attack_counts.index, autopct = '%1.1f%%', textprops={'fontsize': 6})
plt.title('Distribution of Attack Types')
     
legend_labels = list(attack_counts.index)

legend_labels = [
    "Other (Brute Force, Web Attack, Bot)"
    if label == "Other" else label
    for label in legend_labels
]

plt.legend(legend_labels, loc='best', fontsize=8)
plt.show()

attack_counts = attack_counts.drop('Other', errors='ignore')
# Prepare table
table = pd.DataFrame({
    'Count': df['Label'].value_counts(),
    'Percentage (%)': percentages
})

print(table)
```

# 1.0 Separate for every attack types (70% benign, 30% attack)

```python
### Calculating number of benign needed
```

```python
# attack counts (excluding benign)
attack_counts = {
    "DoS": 193745,
    "DDoS": 128016,
    "Port Scan": 90819,
    "Brute Force": 9152,
    "Web Attack": 2143,
    "Bot": 1953
}

ratio = 70 / 30  # = 7/3

benign_needed = {
    attack: int(count * ratio)
    for attack, count in attack_counts.items()
}

benign_needed

benign_df = df[df["Label"] == "BENIGN"]
for attack, count in attack_counts.items():

    # 1. Extract all rows of this attack type
    attack_df = df[df["Label"] == attack]

    # 2. Compute how many benign samples needed
    benign_needed = int(count * ratio)

    # 3. Random sample benign flows
    benign_sample = benign_df.sample(n=benign_needed, replace=False, random_state=42)

    # 4. Combine attack + benign
    combined = pd.concat([attack_df, benign_sample])

    # 5. Shuffle rows (important for ML)
    combined = combined.sample(frac=1, random_state=42).reset_index(drop=True)

    # 6. Save to CSV
    out_path = f"{'/home/wahba/Documents/nids5/test/dataset/1_Attacks_70_30'}/{attack.replace(' ', '_')}.csv"
    combined.to_csv(out_path, index=False)

    print(f"Saved {attack}  {out_path}")
```
